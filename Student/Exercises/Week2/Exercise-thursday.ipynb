```python
import numpy as np
import plotly.graph_objects as go
from IPython.display import display

# نقاط ابتدائية
x = np.array([1.2, 2.8, 4.0, 6.5, 8.2], dtype=float)
y = np.array([2.1, 3.0, 4.4, 5.6, 7.3], dtype=float)

```

```python
def fit_line(x, y):
    if len(x) < 2:
        m = 0.0
        b = float(y[0]) if len(y) else 0.0
    else:
        m, b = np.polyfit(x, y, 1)
    return m, b

def line_points(x, y, xmin=0, xmax=10):
    m, b = fit_line(x, y)
    xs = np.array([xmin, xmax], dtype=float)
    ys = m * xs + b
    return xs, ys, m, b

xs, ys, m, b = line_points(x, y)

fig = go.FigureWidget()

fig.add_scatter(
    x=x,
    y=y,
    mode="markers+text",
    text=[f"({xi:.1f}, {yi:.1f})" for xi, yi in zip(x, y)],
    textposition="top right",
    marker=dict(size=12),
    name="Points"
)

fig.add_scatter(
    x=xs,
    y=ys,
    mode="lines",
    name="Linear regression"
)
fig.update_layout(
    title=f"Interactive Linear Regression | slope={m:.3f}, intercept={b:.3f}",
    dragmode="pan",
    width=900,
    height=550,
    xaxis=dict(range=[0, 10], title="X"),
    yaxis=dict(range=[0, 10], title="Y"),
)

display(fig)

drag_index = {"i": None}

def refresh():
    global x, y
    xs, ys, m, b = line_points(x, y)
    with fig.batch_update():
        fig.data[0].x = x
        fig.data[0].y = y
        fig.data[0].text = [f"({xi:.1f}, {yi:.1f})" for xi, yi in zip(x, y)]
        fig.data[1].x = xs
        fig.data[1].y = ys
        fig.layout.title = f"Interactive Linear Regression | slope={m:.3f}, intercept={b:.3f}"

# التقاط النقطة المختارة
# التقاط النقطة المختارة
def on_click(trace, points, state):
    if points.point_inds:
        drag_index["i"] = points.point_inds[0]
        update_sliders_from_selected()

fig.data[0].on_click(on_click)
# عناصر تحكم لتحديث مكان النقطة المختارة
import ipywidgets as widgets

x_slider = widgets.FloatSlider(description='X', min=0, max=10, step=0.1, value=0)
y_slider = widgets.FloatSlider(description='Y', min=0, max=10, step=0.1, value=0)
info = widgets.HTML("<b>اختر نقطة أولاً بالضغط عليها، ثم حرّك السلايدر.</b>")

def on_slider_change(change):
    i = drag_index["i"]
    if i is None:
        return
    x[i] = x_slider.value
    y[i] = y_slider.value
    refresh()

x_slider.observe(on_slider_change, names='value')
y_slider.observe(on_slider_change, names='value')

def update_sliders_from_selected():
    i = drag_index["i"]
    if i is not None:
        x_slider.value = float(x[i])
        y_slider.value = float(y[i])

display(info, x_slider, y_slider)

# تحديث السلايدر عند الضغط على نقطة
def on_click_update(trace, points, state):
    if points.point_inds:
        drag_index["i"] = points.point_inds[0]
        update_sliders_from_selected()

```

```python
from IPython.display import HTML, display

display(HTML("""
<div style="font-family: Arial, sans-serif; max-width: 980px;">
  <h2 style="margin-bottom:8px;">Interactive Linear Regression</h2>
  <div style="margin-bottom:10px; color:#444;">
    اسحب أي نقطة بالماوس، ودبل كليك داخل الرسم لإضافة نقطة جديدة.
  </div>

  <div style="display:flex; gap:16px; margin-bottom:12px;">
    <div style="padding:10px 14px; background:#f3f4f6; border-radius:12px;">
      <div style="font-size:12px; color:#666;">Slope</div>
      <div id="slopeBox" style="font-size:22px; font-weight:700;">0.000</div>
    </div>
    <div style="padding:10px 14px; background:#f3f4f6; border-radius:12px;">
      <div style="font-size:12px; color:#666;">Intercept</div>
      <div id="interceptBox" style="font-size:22px; font-weight:700;">0.000</div>
    </div>
    <div style="padding:10px 14px; background:#f3f4f6; border-radius:12px;">
      <div style="font-size:12px; color:#666;">Points</div>
      <div id="countBox" style="font-size:22px; font-weight:700;">0</div>
    </div>
  </div>

  <svg id="chart" width="920" height="560" style="border:1px solid #ddd; border-radius:16px; background:white; touch-action:none;"></svg>
</div>

<script>
(() => {
  const svg = document.getElementById("chart");
  const width = 920;
  const height = 560;
  const pad = 55;
  const xMin = 0, xMax = 10;
  const yMin = 0, yMax = 10;

  let points = [
    {id: 1, x: 1.2, y: 2.1},
    {id: 2, x: 2.8, y: 3.0},
    {id: 3, x: 4.0, y: 4.4},
    {id: 4, x: 6.5, y: 5.6},
    {id: 5, x: 8.2, y: 7.3}
  ];

  let draggingId = null;

  const slopeBox = document.getElementById("slopeBox");
  const interceptBox = document.getElementById("interceptBox");
  const countBox = document.getElementById("countBox");

  function clamp(v, min, max) {
    return Math.max(min, Math.min(max, v));
  }

  function xToPx(x) {
    return pad + ((x - xMin) / (xMax - xMin)) * (width - 2 * pad);
  }

  function yToPx(y) {
    return height - pad - ((y - yMin) / (yMax - yMin)) * (height - 2 * pad);
  }

  function pxToX(px) {
    return xMin + ((px - pad) / (width - 2 * pad)) * (xMax - xMin);
  }

  function pxToY(py) {
    return yMin + ((height - pad - py) / (height - 2 * pad)) * (yMax - yMin);
  }

  function fitLine(data) {
    const n = data.length;
    if (n < 2) {
      return {m: 0, b: n ? data[0].y : 0};
    }

    let sumX = 0, sumY = 0;
    for (const p of data) {
      sumX += p.x;
      sumY += p.y;
    }

    const meanX = sumX / n;
    const meanY = sumY / n;

    let num = 0, den = 0;
    for (const p of data) {
      num += (p.x - meanX) * (p.y - meanY);
      den += (p.x - meanX) ** 2;
    }

    const m = den === 0 ? 0 : num / den;
    const b = meanY - m * meanX;
    return {m, b};
  }

  function createSvgEl(tag, attrs = {}) {
    const el = document.createElementNS("http://www.w3.org/2000/svg", tag);
    for (const [k, v] of Object.entries(attrs)) {
      el.setAttribute(k, v);
    }
    return el;
  }

  function getMousePos(evt) {
    const rect = svg.getBoundingClientRect();
    return {
      x: evt.clientX - rect.left,
      y: evt.clientY - rect.top
    };
  }

  function render() {
    svg.innerHTML = "";

    // Grid
    const gridVals = [0, 2, 4, 6, 8, 10];
    for (const v of gridVals) {
      svg.appendChild(createSvgEl("line", {
        x1: xToPx(v), y1: pad,
        x2: xToPx(v), y2: height - pad,
        stroke: "#e5e7eb", "stroke-width": 1
      }));
      svg.appendChild(createSvgEl("text", {
        x: xToPx(v), y: height - pad + 24,
        "text-anchor": "middle",
        fill: "#666", "font-size": 12
      })).textContent = v;

      svg.appendChild(createSvgEl("line", {
        x1: pad, y1: yToPx(v),
        x2: width - pad, y2: yToPx(v),
        stroke: "#e5e7eb", "stroke-width": 1
      }));
      svg.appendChild(createSvgEl("text", {
        x: pad - 14, y: yToPx(v) + 4,
        "text-anchor": "end",
        fill: "#666", "font-size": 12
      })).textContent = v;
    }

    // Axes
    svg.appendChild(createSvgEl("line", {
      x1: pad, y1: height - pad,
      x2: width - pad, y2: height - pad,
      stroke: "#111827", "stroke-width": 2
    }));

    svg.appendChild(createSvgEl("line", {
      x1: pad, y1: pad,
      x2: pad, y2: height - pad,
      stroke: "#111827", "stroke-width": 2
    }));

    svg.appendChild(createSvgEl("text", {
      x: width / 2, y: height - 12,
      "text-anchor": "middle",
      fill: "#444", "font-size": 14
    })).textContent = "X";

    const yLabel = createSvgEl("text", {
      x: 20, y: height / 2,
      "text-anchor": "middle",
      fill: "#444", "font-size": 14,
      transform: `rotate(-90 20 ${height/2})`
    });
    yLabel.textContent = "Y";
    svg.appendChild(yLabel);

    // Regression line
    const {m, b} = fitLine(points);
    const x1 = xMin, y1 = m * x1 + b;
    const x2 = xMax, y2 = m * x2 + b;

    svg.appendChild(createSvgEl("line", {
      x1: xToPx(x1), y1: yToPx(y1),
      x2: xToPx(x2), y2: yToPx(y2),
      stroke: "#111827", "stroke-width": 3
    }));

    // Points
    for (const p of points) {
      const g = createSvgEl("g");

      const circle = createSvgEl("circle", {
        cx: xToPx(p.x),
        cy: yToPx(p.y),
        r: 8,
        fill: "#2563eb",
        cursor: "grab",
        "data-id": p.id
      });

      circle.addEventListener("pointerdown", (evt) => {
        evt.stopPropagation();
        draggingId = p.id;
        circle.setPointerCapture(evt.pointerId);
      });

      const label = createSvgEl("text", {
        x: xToPx(p.x) + 10,
        y: yToPx(p.y) - 10,
        fill: "#475569",
        "font-size": 12
      });
      label.textContent = `(${p.x.toFixed(1)}, ${p.y.toFixed(1)})`;

      g.appendChild(circle);
      g.appendChild(label);
      svg.appendChild(g);
    }

    slopeBox.textContent = m.toFixed(3);
    interceptBox.textContent = b.toFixed(3);
    countBox.textContent = points.length;
  }

  svg.addEventListener("pointermove", (evt) => {
    if (draggingId === null) return;

    const pos = getMousePos(evt);
    const newX = clamp(pxToX(pos.x), xMin, xMax);
    const newY = clamp(pxToY(pos.y), yMin, yMax);

    points = points.map(p =>
      p.id === draggingId
        ? {...p, x: Number(newX.toFixed(2)), y: Number(newY.toFixed(2))}
        : p
    );

    render();
  });

  svg.addEventListener("pointerup", () => {
    draggingId = null;
  });

  svg.addEventListener("pointerleave", () => {
    draggingId = null;
  });

  svg.addEventListener("dblclick", (evt) => {
    const pos = getMousePos(evt);
    const x = clamp(pxToX(pos.x), xMin, xMax);
    const y = clamp(pxToY(pos.y), yMin, yMax);

    points.push({
      id: Date.now() + Math.random(),
      x: Number(x.toFixed(2)),
      y: Number(y.toFixed(2))
    });

    render();
  });

  render();
})();
</script>
"""))
```

```python
import numpy as np
from scipy import stats

battery_life = [9.2, 9.5, 9.8, 9.1, 9.0, 9.7, 9.3, 9.4, 9.6, 9.2]

# اختبار الفرضية
t_stat, p_value = stats.ttest_1samp(battery_life, 10)

print("t statistic:", t_stat)
print("p value:", p_value)

if p_value < 0.05:
    print("Reject H0")
else:
    print("Fail to reject H0")
```

```python
online = [78, 85, 82, 88, 90, 84, 79, 81]
in_person = [92, 88, 91, 87, 95, 89, 93, 90]

t_stat, p_value = stats.ttest_ind(online, in_person)

print("t statistic:", t_stat)
print("p value:", p_value)

if p_value < 0.05:
    print("Reject H0")
else:
    print("Fail to reject H0")
```
